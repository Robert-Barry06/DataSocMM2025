# Astronomy Logic Puzzle: Greek Letters, Symbols & Phonetics

Solve each riddle by translating the Greek letters and symbols to **English letters**. Use arrows to pick letters:

- $\rightarrow$ → last letter of the symbol’s sound  
- $\leftarrow$ → first letter of the symbol’s sound  
- $\leftrightarrow$ → first and last letters of the symbol’s sound
